```python
import pandas as pd
from tables import *
import numpy as np
from tqdm import tqdm_notebook as tqdm
import csv
```

```python
path = '/home/User1/data/self_citations/'
```

```python
articles = {}
with open(path+'article.txt','r') as f:
    reader = csv.reader(f,delimiter='\t')
    next(reader)
    for line in reader:
        articles[int(line[0])] = int(line[2])
```

```python
authors_info = {}
with open(path+'authors_info.csv','r') as f:
    reader = csv.reader(f,delimiter='\t')
    next(reader)
    for line in reader:
        authors_info[int(line[0])] = [line[1],line[2]]
```

```python
disciplines = pd.read_csv(path+'liste_discipline.txt',encoding='latin_1',sep='\t')
list_disciplines = disciplines['EDiscipline']
```

```python
nb_disciplines = len(list_disciplines)
citations = {}
references = {}
for i in range(nb_disciplines):
    disc = list_disciplines[i]
    #row:stats type, cols: years, dep: 1:years since first, 2: year of pub
    #row 1: self %, row 2: self count, row 3: coauthors %,  row 4: co count
    citations[disc] = np.zeros((4,121,2))
    references[disc] = np.zeros((4,121,2))
```

```python
h5file = open_file(path + 'self_citations.h5', mode="r", title="Self Citations")
table = h5file.root.counts.self_cits
```

```python
min_year = 1980
```

```python
nb_lines = len(table)
for i in tqdm(range(nb_lines)):
    line = table[i]
    art_year = articles[line['ID_Art']]
    if art_year >= min_year :
        info = authors_info[line['Cluster_ID']]
        disc = info[0]
        first_year = int(info[1])
        diff_year = art_year - first_year
        total_cits = line['direct_cits'] + line['indirect_cits'] + line['other_cits']
        total_refs = line['direct_refs'] + line['indirect_refs'] + line['other_refs']
        idx_art_year = art_year-1898

        if total_cits > 0:
            citations[disc][0,diff_year,0] += line['direct_cits']/total_cits
            citations[disc][1,diff_year,0] += 1
            citations[disc][2,diff_year,0] += line['indirect_cits']/total_cits
            citations[disc][3,diff_year,0] += 1

            citations[disc][0,idx_art_year,1] += line['direct_cits']/total_cits
            citations[disc][1,idx_art_year,1] += 1
            citations[disc][2,idx_art_year,1] += line['indirect_cits']/total_cits
            citations[disc][3,idx_art_year,1] += 1

        if total_refs > 0:
            references[disc][0,diff_year,0] += line['direct_refs']/total_refs
            references[disc][1,diff_year,0] += 1
            references[disc][2,diff_year,0] += line['indirect_refs']/total_refs
            references[disc][3,diff_year,0] += 1

            references[disc][0,idx_art_year,1] += line['direct_refs']/total_refs
            references[disc][1,idx_art_year,1] += 1
            references[disc][2,idx_art_year,1] += line['indirect_refs']/total_refs
            references[disc][3,idx_art_year,1] += 1
```

```python
import pickle
with open(path+'complied_stats_citations.p','wb') as f:
    pickle.dump([citations,references,list_disciplines],f)
```

```python
from matplotlib import pyplot as plt
```

```python
axis_year = np.arange(min_year,2020)
nb_years = len(axis_year)
```

```python
for i in range(nb_disciplines):
    disc = list_disciplines[i]
    cits = citations[disc]
    refs = references[disc]
    
    idx_data = np.where(cits[1,:,0])[0]
    perc_self_cits = np.zeros(121)
    perc_self_cits[idx_data] = np.divide(cits[0,idx_data,0],cits[1,idx_data,0])*100
    idx_data = np.where(cits[3,:,0])[0]
    perc_co_cits = np.zeros(121)
    perc_co_cits[idx_data] = np.divide(cits[2,idx_data,0],cits[3,idx_data,0])*100
    
    idx_data = np.where(refs[1,:,0])[0]
    perc_self_refs = np.zeros(121)
    perc_self_refs[idx_data] = np.divide(refs[0,idx_data,0],refs[1,idx_data,0])*100
    idx_data = np.where(refs[3,:,0])[0]
    perc_co_refs = np.zeros(121)
    perc_co_refs[idx_data] = np.divide(refs[2,idx_data,0],refs[3,idx_data,0])*100
    
    idx_data = np.where(cits[1,:,1])[0]
    perc_self_cits_year = np.zeros(121)
    perc_self_cits_year[idx_data] = np.divide(cits[0,idx_data,1],cits[1,idx_data,1])*100
    idx_data = np.where(cits[3,:,1])[0]
    perc_co_cits_year = np.zeros(121)
    perc_co_cits_year[idx_data] = np.divide(cits[2,idx_data,1],cits[3,idx_data,1])*100
    
    idx_data = np.where(refs[1,:,1])[0]
    perc_self_refs_year = np.zeros(121)
    perc_self_refs_year[idx_data] = np.divide(refs[0,idx_data,1],refs[1,idx_data,1])*100
    idx_data = np.where(refs[3,:,1])[0]
    perc_co_refs_year = np.zeros(121)
    perc_co_refs_year[idx_data] = np.divide(refs[2,idx_data,1],refs[3,idx_data,1])*100
    
    plt.figure(figsize=(15,12))
    plt.subplot(221)
    plt.plot(perc_self_cits[:nb_years],'r',label='% self citations')
    plt.plot(perc_co_cits[:nb_years],'r--',label='% coauthors citations')
    plt.plot(perc_self_refs[:nb_years],'g',label='% self references')
    plt.plot(perc_co_refs[:nb_years],'g--',label='% coauthors references')
    plt.title(disc)
    plt.ylabel('%')
    plt.xlabel('Year since first publication')
    plt.legend()


    plt.subplot(222)
    plt.plot(axis_year,perc_self_cits_year[-nb_years:],'r',label='% self citations')
    plt.plot(axis_year,perc_co_cits_year[-nb_years:],'r--',label='% coauthors citations')
    plt.plot(axis_year,perc_self_refs_year[-nb_years:],'g',label='% self references')
    plt.plot(axis_year,perc_co_refs_year[-nb_years:],'g--',label='% coauthors references')
    plt.title(disc)
    plt.ylabel('%')
    plt.xlabel('Year of publication')
    plt.legend()
    
    plt.subplot(223)
    plt.plot(perc_self_cits[:nb_years]+perc_co_cits[:nb_years],'b',label='% citations')
    plt.plot(perc_self_refs[:nb_years]+perc_co_refs[:nb_years],'orange',label = '% references')

    plt.ylabel('%')
    plt.xlabel('Year since first publication')
    plt.legend()


    plt.subplot(224)
    plt.plot(axis_year,perc_self_cits_year[-nb_years:]+perc_co_cits_year[-nb_years:],'b',label='% citations')
    plt.plot(axis_year,perc_self_refs_year[-nb_years:]+perc_co_refs_year[-nb_years:],'orange',label = '% references')
    plt.ylabel('%')
    plt.xlabel('Year of publication')
    plt.legend()
    
    plt.savefig(path+'figs/' + disc + '.png')
    plt.close()

   
```
