```python
def hIndex(citations):
        """
        :type citations: List[int]
        :rtype: int
        """
        citations = list(citations)
        if not citations:
            return 0
        citations.sort()
        for i in range(1,len(citations)+1)[::-1]:
            if citations[-i] >= i:
                return i
        return 0
```
