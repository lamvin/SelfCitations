```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
path = '/home/User1/data/self_citations/'
```

```python
min_year = 1980
max_year = 2019
nb_years = max_year-min_year
```

```python
df = pd.read_csv(path+'model/authors_infos_model.csv',sep='\t')
```

```python
df['duration'] = df['max_year'] - df['min_year'] + 1
```

```python
df['age'] = 2019-df['min_year']
```

```python
df = df.loc[df['discipline']!='Unknown']
```

```python
df = df.loc[df['min_year']>=min_year]
```

```python
disc_groups = {'Arts and Humanities':['Arts','Humanities'],
              'Natural Sciences and Engineering':['Biology','Physics','Mathematics','Engineering and Technology','Earth and Space','Chemistry'],
              'Social Sciences':['Social Sciences','Psychology','Professional Fields','Health'],
             'Health Sciences':['Biomedical Research','Clinical Medicine']}
```

```python
group_map = {}
for group in disc_groups:
    for disc in disc_groups[group]:
        group_map[disc] = group
```

```python
df['group'] = df.apply(lambda x: group_map[x['discipline']],axis=1)
```

```python
df.head()
```

### Age vs duration

```python
bins_duration = [0,1,3,5,10,15,20,25,30,nb_years]
```

```python
nb_bins_duration = len(bins_duration)-1
```

```python
min_art = 5
df_analysis = df.loc[df['nb_articles']>=min_art]
```

```python
df_analysis['bin_duration'] = pd.cut(df_analysis['duration'],bins=bins_duration,labels=bins_duration[1:])
```

```python
df_stats = df_analysis.groupby(['bin_duration','age','group']).agg({'Cluster_ID':'count'}).reset_index()
```

```python
df_stats.rename(columns={'Cluster_ID':'count'},inplace=True)
```

```python
all_ages = pd.unique(df_stats['age'])
nb_ages = len(all_ages)
```

```python
df_stats['perc_count'] = df_stats['count']/df_stats['sum_age']
```

```python
groups = pd.unique(df_stats['group'])
nb_groups = len(groups)
```

##### Lines

```python
col = plt.cm.rainbow(np.linspace(0,1,nb_bins_duration))   
```

```python
plt.figure(figsize=(15,15))
for i in range(nb_groups):
    group = groups[i]
    df_group = df_stats.loc[df_stats['group']==group]
    #mat = df_group.pivot(index='bin_duration', columns='age', values='perc_count')
    plt.subplot(2,2,i+1)
    for j in range(nb_bins_duration):
        val_bin = bins_duration[j+1]
        if j == 0:
            label = '1'
        else:
            label = '[{},{}]'.format(bins_duration[j]+1,val_bin)
            
        plt.plot(df_group.loc[df_group['bin_duration']==val_bin,'age'],
                 df_group.loc[df_group['bin_duration']==val_bin,'perc_count'],
                 color=col[j],
                 linewidth=2,
                label = label)
    if i == 0:
        plt.legend()
    plt.ylabel('Probability',fontsize=25)
    plt.xlabel('Academic age',fontsize=25)
    plt.tick_params(labelsize=18)
    plt.title(group,fontsize=25)
out_path = path+'results/authors_career/'
plt.show()
if not os.path.exists(out_path):
    os.makedirs(out_path)
plt.savefig(out_path + 'plot_age_duration_perc_minArt{}.png'.format(min_art))
plt.savefig(out_path + 'plot_age_duration_perc_minArt{}.svg'.format(min_art))
plt.close()
```

##### Heatmaps
