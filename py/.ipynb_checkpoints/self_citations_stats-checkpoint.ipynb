```python
class self_cits(IsDescription):
    Cluster_ID = UInt32Col()
    ID_Art = UInt32Col()
    other_refs = UInt16Col()
    direct_refs = UInt16Col()
    indirect_refs = UInt16Col()
    direct_cits = UInt16Col()
    indirect_cits = UInt16Col()
    other_cits = UInt16Col()
    
```

```python
h5file = open_file(path + 'self_citations.h5', mode="w", title="Self Citations")
group = h5file.create_group("/", 'counts')
table = h5file.create_table(group, 'self_cits', self_cits)
h5file.close()
```

```python
def get_cits_stats(inqueue, output):
    ID = authors[i]
    arts_author = set(dict_cluster_ID[ID])
    args = []
    for ID_Art in arts_author:
        authors_art = dict_cluster_art[ID_Art]
        all_arts_from_authors = np.array([])
        for auth_ID in authors_art:
            all_arts_from_authors = np.concatenate([all_arts_from_authors,dict_cluster_ID[auth_ID]])
        all_arts_from_authors = set(all_arts_from_authors)
        other_cits = 0
        direct_cits = 0
        indirect_cits = 0
        other_refs = 0
        direct_refs = 0
        indirect_refs = 0

        if ID_Art in dict_citant:
            references = dict_citant[ID_Art]
            for ref in references:
                if ref in arts_author:
                    direct_refs += 1
                elif ref in all_arts_from_authors:
                    indirect_refs += 1
                else:
                    other_refs += 1
        if ID_Art in dict_cite:
            cits = dict_cite[ID_Art]
            for cit in cits:
                if cit in arts_author:
                    direct_cits += 1
                elif cit in all_arts_from_authors:
                    indirect_cits += 1
                else:
                    other_cits += 1
        art_info = {}
        art_info['ID_Art'] = ID_Art
        art_info['Cluster_ID'] = ID
        art_info['other_cits'] = other_cits
        art_info['direct_cits'] = direct_cits
        art_info['indirect_cits'] = indirect_cits
        art_info['other_refs'] = other_refs
        art_info['direct_refs'] = direct_refs
        art_info['indirect_refs'] = indirect_refs
        args.append(art_info)
    output.put(args)                    

            
```

```python
def handle_output(output):
    start_time = time.time()
    counter = 0
    f = open(path+'progress.txt','w')
    h5file = open_file(path + 'self_citations.h5', mode="a", title="Self Citations")
    table = h5file.root.counts.self_cits
    row = table.row
    while True:
        args = output.get()
        if args:
            for arg in args:
                row['ID_Art'] = arg['ID_Art']
                row['Cluster_ID'] = arg['Cluster_ID']
                row['other_cits'] = arg['other_cits']
                row['direct_cits'] = arg['direct_cits']
                row['indirect_cits'] = arg['indirect_cits']
                row['other_refs'] = arg['other_refs']
                row['direct_refs'] = arg['direct_refs']
                row['indirect_refs'] = arg['indirect_refs']
                row.append()
            counter += 1
            if counter % 1 == 0 :
                elapsed_time = time.time() - start_time
                elapsed_time_h = np.round(elapsed_time/3600,2)
                time_per_art = elapsed_time/counter
                time_left_h = np.round(time_per_art*(nb_authors-counter)/3600,2)
                perc = np.round(counter/nb_authors*100,3)
                f.write('Progress: {} manuscripts, {}%, Time since start: {}, Time left: {}\n'.format(counter,perc,elapsed_time_h,time_left_h))    
                f.flush()
            table.flush()
        else:
            break
    h5file.close()
    f.close()
```

```python
num_processes = 2
sentinel = None
output = mp.Queue()
inqueue = mp.Queue()
jobs = []
proc = mp.Process(target=handle_output, args=(output, ))
proc.start()

for i in range(num_processes):
    p = mp.Process(target=get_cits_stats, args=(inqueue, output))
    jobs.append(p)
    p.start()
authors_idx = np.arange(nb_authors)
np.random.shuffle(authors_idx)
for i in authors_idx:
    inqueue.put(i)
for i in range(num_processes):
    # Send the sentinal to tell Simulation to end
    inqueue.put(sentinel)
for p in jobs:
    p.join()
output.put(None)
proc.join()
```

```python
start_time = time.time()
f = open(path+'progress.txt','w')
for i in range(nb_articles):
    ID_Art = articles[arts_idx[i]]
    authors_art = dict_cluster_art[ID_Art]
    all_arts_from_authors = set()
    for auth_ID in authors_art:
        all_arts_from_authors.update(dict_cluster_ID[auth_ID])
    for ID in authors_art:
        ID = authors[authors_idx[i]]
        arts_author = set(dict_cluster_ID[ID])

        other_cits = 0
        direct_cits = 0
        indirect_cits = 0
        other_refs = 0
        direct_refs = 0
        indirect_refs = 0

        if ID_Art in dict_citant:
            references = dict_citant[ID_Art]
            for ref in references:
                if ref in arts_author:
                    direct_refs += 1
                elif ref in all_arts_from_authors:
                    indirect_refs += 1
                else:
                    other_refs += 1
        if ID_Art in dict_cite:
            cits = dict_cite[ID_Art]
            for cit in cits:
                if cit in arts_author:
                    direct_cits += 1
                elif cit in all_arts_from_authors:
                    indirect_cits += 1
                else:
                    other_cits += 1

        row['ID_Art'] = ID_Art
        row['Cluster_ID'] = ID
        row['other_cits'] = other_cits
        row['direct_cits'] = direct_cits
        row['indirect_cits'] = indirect_cits
        row['other_refs'] = other_refs
        row['direct_refs'] = direct_refs
        row['indirect_refs'] = indirect_refs
        row.append()
    if i % 1000 == 1 :
        elapsed_time = time.time() - start_time
        elapsed_time_h = np.round(elapsed_time/3600,2)
        time_per_art = elapsed_time/i
        time_left_h = np.round(time_per_art*(nb_articles-i)/3600,2)
        perc = np.round(i/nb_articles*100,3)
        f.write('Progress: {} articles, {}%, Time since start: {}, Time left: {}\n'.format(i,perc,elapsed_time_h,time_left_h))    
        f.flush()
    table.flush()
```
