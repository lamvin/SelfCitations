```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from tqdm import tqdm_notebook as tqdm
import csv
from scipy import sparse
import pickle
import os
import seaborn as sns
```

```python
path = '/home/User1/data/self_citations/model/'
```

#### Read output from make_data_authors.R

```python
data = pd.read_csv(path+'data_authors_rank.csv')
```

```python
data.loc[data['gender']=='m','gender'] = 1
data.loc[data['gender']=='f','gender'] = 0
data['gender'] = data['gender'].astype(np.int)
```

```python
year_start = 5
diff_year = 10
year_pred = year_start+diff_year
```

```python
data = data.loc[data['nb_art_'+str(year_start)]>9]
#data = data.loc[data['active_'+str(year_pred)]]
```

```python
data["norm_cits_co_" +str(year_start)] = data["nb_cits_co_"+str(year_start)]/(data["nb_art_"+str(year_start)]-1)
data["norm_cits_self_" +str(year_start)] = data["nb_cits_self_"+str(year_start)]/(data["nb_art_"+str(year_start)]-1)
data["norm_cits_others_" +str(year_start)] = data["nb_cits_others_"+str(year_start)]/(data["nb_art_"+str(year_start)]-1)

data["norm_cits_co_" +str(year_pred)] = data["nb_cits_co_"+str(year_pred)]/(data["nb_art_"+str(year_pred)]-1)
data["norm_cits_self_" +str(year_pred)] = data["nb_cits_self_"+str(year_pred)]/(data["nb_art_"+str(year_pred)]-1)
data["norm_cits_others_" +str(year_pred)] = data["nb_cits_others_"+str(year_pred)]/(data["nb_art_"+str(year_pred)]-1)
```

```python
data.columns
```

```python
#bins = [0,0.5,0.6,0.7,0.8,0.9,0.95,1]
bins = [0,0.2,0.4,0.6,0.8,0.9,0.95,0.99,1]
```

```python
group_var = 'discipline'
groups = pd.unique(data[group_var])
nb_groups = len(groups)
```

```python
for i in range(nb_groups):
    group = groups[i]
    data.loc[data[group_var]==group,'q_self_cits'] = data.loc[data[group_var]==group,
                                                              'norm_cits_self_' + str(year_start)].rank(pct=True)
    data.loc[data[group_var]==group,'qbinned_self'] = pd.cut(data.loc[data[group_var]==group,'q_self_cits'],bins)
    data.loc[data[group_var]==group,'qbinned_self'] = data.loc[data[group_var]==group,'qbinned_self'].apply(lambda x: x.right)
```

```python
len(data)
```

#### With variables start

```python
#Variables start
variables = ['nb_art','norm_cits_self','norm_cits_co','norm_cits_others','nb_co','prop5'] #gender
variables_start = [var + '_' + str(year_start) for var in variables]
#stats_start = {var:['mean','sem'] for var in variables_start}
stats = {var:['mean'] for var in variables_start}
stats['gender'] = ['mean']
plot_vars = []
for var in stats:
    for stat in stats[var]:
        plot_vars.append(var+stat)
```

#### With variables pred

```python

variables = ['nb_art','norm_cits_self','norm_cits_co','norm_cits_others','nb_co','active','prop5'] #gender
#variables = ['nb_art','norm_cits_self','norm_cits_co','norm_cits_others','nb_co','prop5'] #gender
variables_pred = [var + '_' + str(year_pred) for var in variables]
#stats_start = {var:['mean','sem'] for var in variables_start}
stats = {var:['mean'] for var in variables_pred}
stats['gender'] = ['mean']
plot_vars = []
for var in stats:
    for stat in stats[var]:
        plot_vars.append(var+stat)
        
```

```python
name_dir = 'all_start_min3art/'
```

```python
df_stats = data.groupby(['discipline','qbinned_self']).agg(stats).reset_index()
df_stats.columns = ["".join(x) for x in df_stats.columns.ravel()]
```

```python
groups = pd.unique(data['discipline'])
nb_groups = len(groups)
nb_vars = len(variables) + 1
```

```python
plot_vars
```

```python
%matplotlib inline

var = plot_vars[5]
plt.figure(figsize=(12,12))
df = df_stats[[var,group_var,'qbinned_self']]
df = pd.pivot_table(df, index=group_var, columns='qbinned_self')
df.columns = [np.round(float(x[-1]),2) for x in df.columns.ravel()]

sns.heatmap(df, cmap='RdYlGn_r', linewidths=0.5, annot=True)
plt.title(var)
```

```python
for i in range(nb_vars):
    var = plot_vars[i]
    plt.figure(figsize=(12,12))
    df = df_stats[[var,group_var,'qbinned_self']]
    df = pd.pivot_table(df, index=group_var, columns='qbinned_self')
    df.columns = [np.round(float(x[-1]),2) for x in df.columns.ravel()]

    sns.heatmap(df, cmap='RdYlGn_r', linewidths=0.5, annot=True)
    plt.title(var)
    plt.xlabel('Quartile self-citation per article')
    

    out_path = path+'results_model/' + name_dir

    if not os.path.exists(out_path):
        os.makedirs(out_path)
    plt.savefig(out_path+var+'.png')
    #plt.savefig(out_path+'sim_per_prop_cit.svg')
    plt.close()
```
